# Description

Demonstrate the relation of $1/\sigma^{2}$, $\lambda$ range can be well approximated, and max phae.

# Fix max phase

## Plot the approximate Gaussian fucntion with three given $\sigma^{2}$

```python
QC = GFE_circuit()
dy, My = 0.16, 50 # discrete parameters
max_phase = dy*My # max phase is equalt to 8
QC.set_discrete_parameters(dy, My)

# Three given sigma square
sig_squ_1 = 1
sig_squ_2 = 1/2
sig_squ_3 = 1/4
mu = 0

lambda_list = np.linspace(0, 8, 1000)

QC.set_Gaussian_parameters(sig_squ_1, mu)
approximate_GF_list_1 = QC.demo_approximate_GF(lambda_list)
QC.set_Gaussian_parameters(sig_squ_2, mu)
approximate_GF_list_2 = QC.demo_approximate_GF(lambda_list)
QC.set_Gaussian_parameters(sig_squ_3, mu)
approximate_GF_list_3 = QC.demo_approximate_GF(lambda_list)

# compute the absolute value
approximate_GF_list_1 = abs(np.array(approximate_GF_list_1))
approximate_GF_list_2 = abs(np.array(approximate_GF_list_2))
approximate_GF_list_3 = abs(np.array(approximate_GF_list_3))
```

```python
fig = plt.figure(figsize = [6, 4])
plt.plot(lambda_list, approximate_GF_list_1, 'r', label="$1/\sigma^{2}$=1")
plt.plot(lambda_list, approximate_GF_list_2, 'g', label="$1/\sigma^{2}$=2")
plt.plot(lambda_list, approximate_GF_list_3, 'b', label="$1/\sigma^{2}$=4")

plt.yscale('log')

plt.title('Approximate GF')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)

plt.xlabel('$\lambda$',fontsize=16)
plt.ylabel('$G(\lambda)$',fontsize=16)

plt.tight_layout()
plt.show()
```

## Plot the approximate Gaussian function and the exact one

```python
lambda_list = np.linspace(0, 6, 1000)
QC.set_Gaussian_parameters(sig_squ_1, mu)
approximate_GF_list_1 = QC.demo_approximate_GF(lambda_list)
approximate_GF_list_1 = abs(np.array(approximate_GF_list_1))
```

```python
fig = plt.figure(figsize = [6, 4])
plt.plot(lambda_list, approximate_GF_list_1, '--', label="Approximate")
plt.plot(lambda_list, np.exp(-lambda_list**2/sig_squ_1), '-', label="Exact")
plt.vlines(x=sig_squ_1*max_phase/2, ymin=0, ymax=1, linestyles='--', colors='red')
plt.yscale('log')

plt.title('$1/\sigma^{2}$=1')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)

plt.xlabel('$\lambda$',fontsize=16)
plt.ylabel('$G(\lambda)$',fontsize=16)

plt.show()
```

```python
fig = plt.figure(figsize = [6, 4])
plt.plot(lambda_list, approximate_GF_list_2, '--', label="Approximate")
plt.plot(lambda_list, np.exp(-lambda_list**2/sig_squ_2), '-', label="Exact")
plt.vlines(x=sig_squ_2*max_phase/2, ymin=0, ymax=1, linestyles='--', colors='red')
plt.yscale('log')

plt.title('$1/\sigma^{2}$=2')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)

plt.xlabel('$\lambda$',fontsize=16)
plt.ylabel('$G(\lambda)$',fontsize=16)

plt.show()
```

# Fix sigma square

## Plot the approximate Gaussian fucntion with three given max phase

```python
QC = GFE_circuit()

# Gaussian parameter
sig_squ = 1/2
mu = 0
QC.set_Gaussian_parameters(sig_squ, mu)

# Three chosen max phase
dy = 0.16
My1, My2, My3 = 25, 50, 75 # corresponding to max phase 4, 8, and 12
max_phase_1 = dy*My1
max_phase_2 = dy*My2
max_phase_3 = dy*My3

lambda_list = np.linspace(0, 8, 1000)

QC.set_discrete_parameters(dy, My1)
approximate_GF_list_1 = QC.demo_approximate_GF(lambda_list)
QC.set_discrete_parameters(dy, My2)
approximate_GF_list_2 = QC.demo_approximate_GF(lambda_list)
QC.set_discrete_parameters(dy, My3)
approximate_GF_list_3 = QC.demo_approximate_GF(lambda_list)

# compute the absolute value
approximate_GF_list_1 = abs(np.array(approximate_GF_list_1))
approximate_GF_list_2 = abs(np.array(approximate_GF_list_2))
approximate_GF_list_3 = abs(np.array(approximate_GF_list_3))
```

```python
fig = plt.figure(figsize = [6, 4])
plt.plot(lambda_list, approximate_GF_list_1, 'r', label="$\phi_{max}=4$")
plt.plot(lambda_list, approximate_GF_list_2, 'g', label="$\phi_{max}=8$")
plt.plot(lambda_list, approximate_GF_list_3, 'b', label="$\phi_{max}=12$")

plt.yscale('log')

plt.title('Approximate GF')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)

plt.xlabel('$\lambda$',fontsize=16)
plt.ylabel('$G(\lambda)$',fontsize=16)

plt.tight_layout()
plt.show()
```

```python
lambda_list = np.linspace(0, 2, 1000)
QC.set_discrete_parameters(dy, My1)
approximate_GF_list_1 = QC.demo_approximate_GF(lambda_list)
approximate_GF_list_1 = abs(np.array(approximate_GF_list_1))
```
